#### Let's see how much null values that we have got 

```python
train.isnull().sum()
```

- Looks Like We have Null 177 Age Values, 687 Null Cabin Values, 2 Embarked Values

- Let's Replace those values by A Mean Value and Some other Values

```python
train['Age'].fillna((train['Age'].mean()), inplace=True)
train['Embarked'].fillna('S', inplace=True)
train['Cabin'].fillna('C148', inplace=True)
```

```python
train.isnull().sum()
```

#### Nice Look on the up. We have filled Null Values with Mean Values and Some Other Values

## Let's See How Much Of Passengers Have Survived

#### Let's Plot Those Values on a Pie Chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train['Survived'].value_counts().plot.pie(
    shadow=True,
    colors=['mediumseagreen', 'tomato'],
    autopct='%1.1f%%',
    labels=['Died', 'Survived']
)
```

## Let's See How The Passenger Class Have Affected the Survival or the Death of a Passenger

```python
train.groupby('Pclass')['Survived'].mean()
```

### Now Let's See How Much percentage that of 1st 2nd 3rd Class Passengers Have Survived

```python
train.groupby('Pclass')['Survived'].mean().plot.pie(
    autopct='%1.1f%%',
    shadow=True,
    colors=['lightblue', 'tomato', 'mediumseagreen'],
    labels=['1st Class', '2nd Class', '3rd Class']
)
```

## Let's See How The Sex of the Passnger Have Affected the result

```python
train.groupby('Sex')['Survived'].mean()
```

```python
train.groupby('Sex')['Survived'].value_counts()
```

```python
train.groupby('Sex')['Survived'].value_counts().plot.bar()
```

# Here We can See,
- 233 Female Survived While 81 Died

## Let's See How the Passenger Class Have Played a Role in Survival of Both Sex

```python
train.groupby(['Pclass', 'Sex'])['Survived'].mean()
```
